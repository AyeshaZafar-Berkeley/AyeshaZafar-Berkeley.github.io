### Model 1

## Logistical Regression

Logistic regression models the probability of failing an inspection through a linear relationship between predictors and the log-odds of the outcome, assuming independence of observations and a roughly linear effect of features after transformation. Its main advantages are its interpretability, speed, and transparency, policy actors can clearly see which factors increase or decrease inspection risk, and regularisation helps handle correlated predictors. However, it performs poorly when relationships are highly non-linear or involve complex interactions. It is a plausible choice here because food inspection outcomes have meaningful, interpretable covariates (e.g., prior violations, licence type, neighbourhood characteristics), and a simple, explainable baseline helps justify more complex models later.

```python

y_train_flat = y_train.ravel()

# define logistic regression model
logit_model = LogisticRegression(max_iter=500)

# valid hyperparameter grid
logit_param_grid = {
    'penalty': ['l1', 'l2'],              
    'C': [0.01, 0.1, 1, 10, 100],
    'fit_intercept': [True, False],
    'solver': ['liblinear'],              
    'class_weight': [None, 'balanced']
}

# grid search
logit_grid = GridSearchCV(
    estimator=logit_model,
    param_grid=logit_param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

logit_grid.fit(X_train, y_train_flat)

best_logit = logit_grid.best_estimator_
print("\nBest Logistic Regression params:")
print(logit_grid.best_params_)

```

###  Model 2

## Random Forests 

Random forests combine many decision trees built on bootstrap samples with random feature subsets, making minimal assumptions about functional form and allowing the model to capture non-linear relationships and high-order interactions without requiring manual specification. Its pros include strong predictive performance on structured, tabular data; built-in feature importance measures; and robustness to noise, scaling issues, and irrelevant predictors. Its downsides are reduced interpretability relative to logistic regression and potentially high computational cost, especially with large grids. It is well-suited to this project because food inspection outcomes likely depend on complex combinations of factors, such as restaurant type interacting with past violations or location, and a random forest can naturally detect these patterns without overfitting excessively.

```python


# ---- Base RF model ----
rf_base = RandomForestClassifier(
    random_state=42,
    n_jobs=-1
)

# ---- Hyperparameter grid ----
rf_param_grid = {
    'n_estimators': [150],             # fixed (150 is a sweet spot)
    'criterion': ['gini'],             # entropy rarely improves performance
    'max_depth': [None, 12],           # captures simple vs complex trees
    'min_samples_split': [2, 5],       # essential variation
    'min_samples_leaf': [1, 2],        # essential variation
    'max_features': ['sqrt'],          # almost always best for classification
    'class_weight': ['balanced']       # helps with class imbalance
}

# ---- Grid search ----
rf_grid = GridSearchCV(
    estimator=rf_base,
    param_grid=rf_param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

rf_grid.fit(X_train, y_train_flat)

print("\nBest Random Forest params:")
print(rf_grid.best_params_)
print("Best CV Accuracy:", rf_grid.best_score_)

# ---- Fit best model and compute feature importance ----
best_rf = rf_grid.best_estimator_


# Fit best RF on the full training data
best_rf.fit(X_train, y_train_flat)

```

### Model 3

## AdaBoost 

AdaBoost sequentially trains a series of weak learners, typically shallow decision trees, each one focusing more heavily on observations that previous models misclassified. The approach assumes that base learners are weak but better than random and is sensitive to noisy labels because the algorithm keeps re-weighting hard cases. Its strengths include flexibility, the ability to model non-linear decision boundaries, and often strong performance on medium-sized datasets, while its limitations include over-sensitivity to outliers and lower interpretability compared to simpler models. AdaBoost is a plausible model for this project because inspection outcomes contain both easy cases (repeat offenders) and ambiguous cases, and a boosting algorithm that emphasises borderline observations can help identify subtle predictors of risk that simpler models may miss.

```python

y_train_flat = y_train.ravel()

# ---- Base decision tree for AdaBoost ----
base_tree = DecisionTreeClassifier(random_state=42)

# ---- Base AdaBoost model ----
ada_model = AdaBoostClassifier(
    estimator=base_tree,
    random_state=42
)

# ---- Hyperparameter grid ----
ada_param_grid = {
    'estimator__max_depth': [1, 2, 3],
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.5, 1.0]
}

# ---- Grid search ----
ada_grid = GridSearchCV(
    estimator=ada_model,
    param_grid=ada_param_grid,
    cv=5,                        # 5-fold CV = correct for this stage
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

ada_grid.fit(X_train, y_train_flat)

# ---- Best model from CV ----
best_ada = ada_grid.best_estimator_

print("\nBest AdaBoost params:")
print(ada_grid.best_params_)
print("Best CV Accuracy (AdaBoost):", ada_grid.best_score_)

```

### Validation Metrics

Be sure to explain which of these metrics you would want to prioritize when conducting predictive auditing in this context and why.

**Hint**: Try writing a for loop to use [`cross_val_score()`](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html) to check for accuracy, precision, recall and f1 across all of your models.

```python


y_train_flat = y_train.ravel()


models = {
    "Logistic Regression": best_logit,
    "Random Forest": best_rf,
    "AdaBoost": best_ada
}

# metrics 
scoring_metrics = {
    "accuracy": "accuracy",
    "precision": make_scorer(precision_score, zero_division=0),
    "recall": make_scorer(recall_score, zero_division=0),
    "f1": make_scorer(f1_score, zero_division=0)
}

results = []

for model_name, model in models.items():
    row = {'Model': model_name}
    for metric_name, scorer in scoring_metrics.items():
        score = cross_val_score(
            model,
            X_train,
            y_train_flat,
            cv=5,
            scoring=scorer,
            n_jobs=1
        ).mean()
        row[metric_name] = score
    results.append(row)

metrics_df = pd.DataFrame(results)
print(metrics_df)

```

## Explaination 

In this context, I would prioritise recall and F1 for failing inspections over raw accuracy, because the social cost of missing a dangerous resturant is highwe than the cost of inspecting a safe one. Accuracy alone could look high even if the model is missing many truly unsafe establishements, given class imbalance.


All three models, Logistic Regression, Random Forest, and AdaBoost, achieve accuracy values of roughly 0.80. While this may appear strong, accuracy is not informative in this context because the dataset is highly imbalanced, with the majority of establishments passing their inspections. A model that simply predicts "pass" for every observation would also achieve approximately 80% accuracy. Therefore, accuracy masks substantial weaknesses in identifying the minority class (failed inspections).

Precision values range from 0.35 to 0.47, indicating that when the models do predict a failure, they are correct roughly one-third to one-half of the time. However, this is overshadowed by extremely low recall. Recall, ranging from 0.0016 for logistic regression to about 0.0127 for the random forest, shows that the models are capturing far less than 2% of true failing establishments. This means the models are overwhelmingly predicting "pass," missing almost all unsafe establishments. Consequently, F1 scores are also extremely low (0.0030.024), reflecting this poor balance between precision and recall.

Taken together, these results indicate that the models perform poorly at identifying risky establishments, despite superficially high accuracy. This outcome reflects the severe class imbalance in the dataset and highlights why metrics like recall and F1 are more useful for public health and regulatory settings: missing hazardous establishments (false negatives) is far more costly than inspecting safe ones. To improve model performance, techniques such as threshold tuning, class weighting, or resampling are needed to meaningfully increase recall on the fail class.

## 5. Discussion Questions

1. Why do we need metrics beyond accuracy when using machine learning in the social sciences and public policy?


**YOUR ANSWER HERE**...

Accuracy is insufficient in social science and public policy contexts because many policy-relevant outcomes are low-prevalence events, meaning a model can achieve high accuracy while failing to identify the small subset of cases that matter most. This makes accuracy a poor indicator of substantive performance. Instead, metrics such as precision, recall, and F1 are necessary to evaluate how effectively a model detects high-risk or high-cost outcomes and to assess the trade-offs between false positives and false negatives. These metrics provide a more meaningful basis for policy decisions, where the consequences of misclassification, not overall correctness, are paramount.

In the food inspection context, accuracy can be especially misleading because the overwhelming majority of establishments pass their inspections. A model that predicts pass for every business would score high on accuracy yet fail to identify nearly all unsafe establishments. Metrics such as recall and F1 are therefore essential, as they capture the models ability to detect the minority of cases that pose genuine public-health risks and that regulators most need to identify.

2. Imagine that establishments learned about the algorithm being used to determine who gets audited and they started adjusting their behavior (and changing certain key features about themselves that were important for the prediction) to avoid detection. How could policymakers address this interplay between algorithmic decision-making and real-world behavior?

**YOUR ANSWER HERE**...

Policymakers can address this dynamic by regularly retraining and updating the algorithm so that its decision rules evolve and cannot be easily gamed. They can also incorporate a mix of interpretable and non-interpretable features, including some that are difficult for establishments to manipulate (e.g., historical violations or environmental conditions). Finally, policymakers can adopt a hybrid inspection strategy, combining model-driven targeting with a fixed proportion of random audits, which preserves deterrence by making avoidance strategies less reliable and ensures that behavior does not drift in response to the algorithm.
